```python
# Polymorphism

# https://www.edureka.co/blog/polymorphism-in-python/#:~:text=Polymorphism%20in%20python%20defines%20methods,inherited%20from%20the%20parent%20class.

# It means that the same function name can be used for different types.

# You can use different function, class methods or objects to define polymorphism. 
# Polymorphism with Function and Objects

# class Carot():
#     def type(self):
#         print("Vegetable")
#     def color(self):
#         print("RED")
# class Banana():
#     def type(self):
#         print("fruit")
#     def color(self):
#         print("yellow")
        
# def fun(ob):
#     ob.type()
#     ob.color()
# ob_carot=Carot()
# ob_banana=Banana()

# fun(ob_carot)
# fun(ob_banana)

# class Tomato(): 
#      def type(self): 
#        print("Vegetable") 
#      def color(self):
#        print("Red") 
# class Apple(): 
#      def type(self): 
#        print("Fruit") 
#      def color(self): 
#        print("Red") 
      
# def func(obj): 
#        obj.type() 
#        obj.color()
        
# obj_tomato = Tomato() 
# obj_apple = Apple() 
# func(obj_tomato) 
# func(obj_apple)
    
# Polymorphism with Class Methods

# class Tomato():
#     def type(self):
#         print("Vegetable") 
#     def color(self):
#         print("Red") 
# class Apple():
#     def type(self): 
#             print("Fruit") 
#     def color(self):
#             print("Red") 
# obj_tomato = Tomato() 
# obj_apple = Apple()    

# for d in (obj_tomato,obj_apple):
#     d.type()
#     d.color()

# Polymorphism with inheritance (Method over loading)

class Bird:
     def intro(self):
       print("There are different types of birds")
 
     def flight(self):
       print("Most of the birds can fly but some cannot")
 
class parrot(Bird):
     def flight(self):
       print("Parrots can fly")
 
class penguin(Bird):
     def flight(self):
       print("Penguins do not fly")
 
obj_bird = Bird()
obj_parr = parrot()
obj_peng = penguin()
 
obj_bird.intro()
obj_bird.flight()
 
obj_parr.intro()
obj_parr.flight()
 
obj_peng.intro()
obj_peng.flight()

```

```python
# Abstraction 
# Abstraction in OOP is a process of hiding the real implementation of the method by only showing a method signature to the use
'''
To create abstract class first we need to import abc module.After that we need to 
import ABC class as base class for abstract clss.To define any method as abstract method we need to give @abstractmethod0 keywod
above our abstractmethod
'''
# To create object for child class we should create all abstract methods which are created in abstract class

import abc
from abc import ABC,abstractmethod
class Ab_class(ABC):
    def sample(self):
        print("Normal method")
    @abstractmethod
    def sample1(self):
        print("Abastractmethod")

class Main(Ab_class):
    def sample1(self):
        print("Main logic")
ab_ob=Main()
ab_ob.sample1()



class Arith(ABC):
    def __init__(self,a):
        self.a=a
    @abstractmethod    
    def add(self):
        pass
    @abstractmethod    
    def sub(self):
        pass
    
class out(Arith):
    def sub(self):
        print(self.a-12)
    def add(self):
        print(self.a+3)
        
d=out(4)
d.add()
d.sub()
```
